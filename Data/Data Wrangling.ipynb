# Data Preparation

## 1) Accounting for Stock Splits
Stock splits cause an increase in the supply of the stock while the company value remains the same. Typically this is done 2 to 1. Meaning, if you hold 10 stocks of AAPL priced at 150 each, it will become 20 stocks priced at 75 each. Unaccounted for, it would lead to a drastic decrease.

```python
# Set up index
index = pd.date_range(start='1999-01-01', end='2019-01-01')
stock_splits = pd.DataFrame(index=index)

# Get split ratios one company at a time
for i in sptickers:
    split = quandl.get_table('WIKI/PRICES', ticker = i, 
                            qopts = { 'columns': ['ticker', 'date', 'split_ratio'] }, 
                            date = { 'gte': '1999-01-01', 'lte': '2019-01-01' }, 
                            paginate=True)
    # Join data to index
    new_splits = split.set_index('date')
    clean_splits = new_splits.pivot(columns='ticker')
    stock_splits = stock_splits.join(clean_splits, how="left")
```

```python
# Clean up column names
stock_splits.columns = [i[1] for i in stock_splits.columns]
```

```python
# Remove closed market days, and account for NaNs
stock_splits = stock_splits.dropna(axis=0, how='all')
stock_splits = stock_splits.apply(pd.to_numeric)
stock_splits = stock_splits.fillna(1)
```

```python
# Function to create a dataframe of multiplicative factors
def split_factor(split_ratio):
    series = np.asarray(split_ratio)
    split_index = []
    # Get indices where ratio isn't one
    for i in range(len(series)):  
        if(series[i] != 1):
            split_index.append(i)
    # Start at the last value
    split_index.reverse()
    # Multiply all values after the index by the ratio
    for j in split_index:
        series[j+1:len(series)] = series[j+1:len(series)]*series[j]
    return(series)
```

```python
# Use the above function
stock_splits.apply(split_factor)
```

```python
# Fill NaNs and multiply to control for splits
sp500 = sp500.fillna(0)
sp500_unsplit = sp500.multiply(stock_splits)
```

## 2) Log-Returns
Most analyses on stock returns are done in log-returns because of the additive property they have. The sum of 10 consecutive log-returns is the same as the log-return between the first and last of those 10 periods.

```python
# Define a function to keep things clean
def log_return(stock):
    log_ret = np.log(stock) - np.log(stock.shift(1))
    return(log_ret)

# Apply above function across all columns
sp500_lr = sp500_unsplit.apply(log_return)
sp500_lr.head()
```
