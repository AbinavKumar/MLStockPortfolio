For simplicity, I use Quandl's WIKI Price [database](https://www.quandl.com/databases/WIKIP/documentation). Note that the database is not completely up-to-date.

```python
# Create index for data frame. Dates from 1999-01-01 to 2019-01-01.

index = pd.date_range(start='1999-01-01', end='2019-01-01')
sp500 = pd.DataFrame(index=index)

# Because I can't make too many requests at once, I run a loop through the tickers
for i in sptickers:
    data = quandl.get_table('WIKI/PRICES', ticker = i, 
                        qopts = { 'columns': ['ticker', 'date', 'adj_close'] }, 
                        date = { 'gte': '1999-01-01', 'lte': '2019-01-01' }, 
                        paginate=True)
    
    # Set the row index to date and fill in the ticker
    new = data.set_index('date')
    clean_data = new.pivot(columns='ticker')
    sp500 = sp500.join(clean_data, how="left")
    # Counter for the impatient
    print(sptickers.index(i))
    time.sleep(.1)
```

```python
# Drop all observations where the market is not open
sp500v2 = sp500.dropna(axis=0,how='all')
# Clean up the column names
sp500v2.columns = [i[1] for i in sp500v2.columns]
# Save file
sp500v2.to_csv(r"/Users/abikumar/Desktop/Capstone1/SP500.csv")
```

# S&P 500 Indicators
Let's just see what kind of data we can get from SHARADAR. I'll test it using Apple (AAPL).

```python
data = quandl.get_table('SHARADAR/SF1', ticker = 'AAPL')
data
```

It seems that, without paying, I can get up to 8 Most-Recently Reported Yearly Data (MRY). The documentation defines Most-Recently reported as "the most recently reported data for thtat reporting period." 

```python
# Create empty dataframe to store results
indicators = pd.DataFrame(columns = data.columns)

# Go through all tickers and get the indicators
for ticker in sptickers:
    data = quandl.get_table('SHARADAR/SF1', ticker = ticker)
    
    indicators = indicators.append(data)
    
    print(sptickers.index(ticker))
    time.sleep(.1)
```
